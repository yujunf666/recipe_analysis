```python
from dsc80_utils import *
import pandas as pd
import numpy as np
from pathlib import Path
import re
import requests
import time
import otter
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
df_interactions = pd.read_csv('RAW_interactions.csv')
```

```python
df_recipes = pd.read_csv('RAW_recipes.csv')
```

```python
processed_recipe = df_recipes.merge(df_interactions, left_on='id', right_on='recipe_id', how = 'left')
```

```python
processed_recipe['rating'] = processed_recipe['rating'].replace(0, np.nan)
```

```python
avg_rating = processed_recipe.groupby('id').mean()['rating']
df_avg_rating = avg_rating.to_frame()
df_avg_rating = df_avg_rating.rename(columns = {'rating' : 'avg_rating'})
```

```python
clean_df = processed_recipe.merge(df_avg_rating, on='id', how = 'left')
```

```python
clean_df.columns
```

```python
def to_lst(string):
    stripped_string = string.strip('[]')
    my_list = [float(x.strip()) for x in stripped_string.split(',')]
    return my_list
clean_df['lst_nutrition'] = clean_df['nutrition'].apply(to_lst)
clean_df
```

```python
clean_df[['calories', 'total fat (PDV)', 'sugar (PDV)', 'sodium (PDV)', 'protein (PDV)', 'saturated fat (PDV)', 'carbohydrates (PDV)']] = clean_df['lst_nutrition'].apply(pd.Series)
```

```python
def diff_level(step):
    if step <= 6:
        return "Easy"
    elif step <= 12:
        return "Med_1"
    elif step <= 18:
        return "Med_2"
    elif step <= 24:
        return "Med_3"
    else:
        return "Hard"
```

```python
clean_df['diff_level'] = clean_df['n_steps'].apply(diff_level)
```

```python
# Hypothesis Test
```

```python
clean_df['n_ingredients'].max()
```

```python
ingredients_df = clean_df.copy()
ingredients_df['diverse'] = ingredients_df['n_ingredients'] >= 10
ingredients_df = ingredients_df[['n_ingredients', 'rating', 'diverse']]
ingredients_df
```

```python
def diff_of_mean(df):
    not_diverse = df[df['diverse'] == 0]['rating'].mean()
    diverse = df[df['diverse'] == 1]['rating'].mean()
    diff = not_diverse - diverse
    return diff
```

```python
def simulation():
    shuffled = ingredients_df.copy()
    shuffled['diverse'] = np.random.permutation(shuffled['diverse'])
    result = diff_of_mean(shuffled)
    return result
```

```python
observed_diff = diff_of_mean(ingredients_df)
diffs = []
for i in range(1000):
    diff = simulation()
    diffs.append(diff)
p_value = (diffs >= observed_diff).mean()
```

```python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def multi_type_pipeline(data):
    
    all_transform = ColumnTransformer(transformers=[
        ('Std1', StandardScaler(), ['rating']),
        ('Std2', StandardScaler(), ['minutes']),
        ('One_hot_code', OneHotEncoder(), ['diff_level'])
    ])
    
    pl_2 = Pipeline([
        ('Preprocessor', all_transform),
        ('Linear_regression', LinearRegression())
    ])
    
    x = data[['rating', 'minutes', 'diff_level']]
    y = data['n_ingredients']
    
    pl_2.fit(x, y)
    prediction = pl_2.predict(x)
    r2 = pl_2.score(x, y)
    rmse = np.sqrt(mean_squared_error(y, prediction))
    
    return (rmse, r2)
predict_df = clean_df[['rating', 'minutes', 'diff_level', 'n_ingredients']]
predict_df.dropna(inplace=True)
multi_type_pipeline(predict_df)
```

```python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import FunctionTransformer, OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

def multi_type_pipeline(data):
    def itself(x):
        return x
    transform_self = FunctionTransformer(itself)
    all_transform = ColumnTransformer(transformers=[
        ('itself1', transform_self, ['sugar (PDV)']),
        ('itself2', transform_self, ['protein (PDV)']),
        ('Std1', StandardScaler(), ['rating']),
        ('Std2', StandardScaler(), ['minutes']),
        ('One_hot_code', OneHotEncoder(), ['diff_level'])
    ])
    
    pl_2 = Pipeline([
        ('Preprocessor', all_transform),
        ('Linear_regression', LinearRegression())
    ])
    
    x = data[['rating', 'minutes', 'diff_level', 'sugar (PDV)', 'protein (PDV)']]
    y = data['n_ingredients']
    
    pl_2.fit(x, y)
    prediction = pl_2.predict(x)
    r2 = pl_2.score(x, y)
    rmse = np.sqrt(mean_squared_error(y, prediction))
    return (rmse, r2)
predict_df = clean_df[['rating', 'minutes', 'diff_level', 'n_ingredients', 'sugar (PDV)', 'protein (PDV)']]
predict_df.dropna(inplace=True)
multi_type_pipeline(predict_df)
```
